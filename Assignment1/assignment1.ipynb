TГӨmГӨn harjoituksen tavoitteena on rakentaa yksinkertainen neuroverkko, joka tunnistaa lajin (Iris) neljГӨn numeerisen piirteen perusteella.
* Verholehden pituus (Sepal Length)
* Verholehden leveys (Sepal Width)
* TerГӨlehden pituus (Petal Length)
* TerГӨlehden leveys (Petal Width)

Harjoituksessa kГӨytГӨn Keras-kirjastoa rakentaakseen mallin, joka luokittelee kukat kolmen eri lajiin.
* Iris-Setosa
* Iris-Versicolor
* Iris-Virginica


```python
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=column_names)

df.head()
```

Neuroverkkoa varten data on valmisteltava:
1. AlkuperГӨinen data on jГӨrjestetty lajeittain., KГӨytГӨn df.sample(frac1), jotta opetus- ja validointidatassa on kaikkia lajeja sekaisin
2. Muutan lajinimet vektoreiksi. Esimerkiksi laji 0 muuttuu

```python
df = df.sample(frac=1).reset_index(drop=True)

X = df.drop('species', axis=1).values
y_raw = df['species'].values

encoder = LabelEncoder()
y_int = encoder.fit_transform(y_raw)
y = to_categorical(y_int)

df.head()
```

TehtГӨvГӨannon mukaisesti rakennan mallin, jossa on:
* Yksi piilokerros
* Ulostulokerros

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(4,)),
    Dense(3, activation='softmax')
])

opt = Adam(learning_rate=0.01)

model.compile(optimizer=opt,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

KГӨytГӨn 'validation_split=0.2' mallin opetukseen, eli jГӨtГӨn 20% datasta mallin testamiseen. Tavoitteena on saavuttaa vГӨhintГӨГӨn 95% tarkkuus.

```python
history = model.fit(X, y,
                    epochs=50,
                    batch_size=5,
                    validation_split=0.2,)
```
